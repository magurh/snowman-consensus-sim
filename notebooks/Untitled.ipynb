```python
import matplotlib.pyplot as plt
import pandas as pd

import os
import sys

module_path = os.path.abspath("../")  # root directory
if module_path not in sys.path:
    sys.path.append(module_path)

data_path = os.path.join(module_path, 'data')
```

```python
file = os.path.join(data_path, 'flare_default_2col.csv')
df = pd.read_csv(file)

df['NumRoundsTree'] = df['NumRoundsTree']/100
df['NumRoundsFlat'] = df['NumRoundsFlat']/100

plt.figure(figsize=(10, 6))
plt.hist(df["NumRoundsTree"], bins=50, alpha=0.6, label="Tree", color='blue')
plt.hist(df["NumRoundsFlat"], bins=75, alpha=0.6, label="Flat", color='red')

# Add labels and title
plt.xlabel("Number of Rounds per Node until Network Convergence")
plt.ylabel("Frequency")
plt.title("Flare Default Parameters - 2 colors")
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()
```

```python
file = os.path.join(data_path, 'flare_multiple_2cols.csv')
df = pd.read_csv(file)

# Create a figure with two subplots side by side
fig, ax = plt.subplots(2, 4, figsize=(24, 8), sharey=True)

# Plot histogram for the first column
for ls in range(4, 9):
    alpha = ls + 7
    ax[0][0].hist(df[str(ls)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    ax[0][1].hist(df[str(ls+8)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    ax[0][2].hist(df[str(ls+2*8)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    ax[0][3].hist(df[str(ls+3*8)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    
ax[0][0].set_title("K = 20, BetaVirtuous = 15, BetaRogue = 20")
ax[0][0].set_xlim([1900, 4500])
ax[0][1].set_title("K = 20, BetaVirtuous = 10, BetaRogue = 13")
ax[0][1].set_xlim([1250, 3500])
ax[0][2].set_title("K = 20, BetaVirtuous = 10, BetaRogue = 11")
ax[0][2].set_xlim([1050, 3000])
ax[0][3].set_title("K = 20, BetaVirtuous = 10, BetaRogue = 16")
ax[0][3].set_xlim([1550, 3500])

for j in range(4):
    ax[0][j].set_ylim([0,340])

for ls in range(1, 4):
    alpha = ls + 5
    ax[1][0].hist(df[str(ls)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    ax[1][1].hist(df[str(ls+8)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    ax[1][2].hist(df[str(ls+2*8)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    ax[1][3].hist(df[str(ls+3*8)], bins=30, alpha=0.7, edgecolor='black', label = f'Alpha = {alpha}')
    
ax[1][0].set_title("K = 10, BetaVirtuous = 15, BetaRogue = 20")
ax[1][0].set_xlim([1900, 4000])
ax[1][1].set_title("K = 10, BetaVirtuous = 10, BetaRogue = 13")
ax[1][1].set_xlim([1250, 3500])
ax[1][2].set_title("K = 10, BetaVirtuous = 10, BetaRogue = 11")
ax[1][2].set_xlim([1050, 3000])
ax[1][3].set_title("K = 10, BetaVirtuous = 10, BetaRogue = 16")
ax[1][3].set_xlim([1550, 3500])


for i in range(2):
    for j in range(4):
        ax[i][j].set_xlabel("Rounds per Node until Convergence")
        ax[i][j].set_ylabel("Frequency")
        ax[i][j].grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
        ax[i][j].legend()
        
# Adjust layout for better spacing
plt.tight_layout()

plt.show()
```
